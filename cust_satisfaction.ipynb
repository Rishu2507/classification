```python
df2 = pd.get_dummies(df,drop_first='if_binary').astype(int)
df2.head()
```

```python
# x and y split
x = df2.drop('satisfaction_satisfied', axis =1)
y = df2[['satisfaction_satisfied']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.isnull().sum
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train ) 
```

```python
print(lg.score(x_train,y_train))
print(lg.score(x_test,y_test))
```

```python
# model testing model evalution
# 0.797236802413273
# 0.7959843621796419
#score down
```

```python
#cofusion matrix
from sklearn.metrics import confusion_matrix, classification_report
```

```python
pred = lg.predict(x_test)
```

```python
pred
```

```python
y_test['prediction'] = pred
```

```python
y_test
```

```python
y_test.head(40)
```

```python
matrix = confusion_matrix(y_test['satisfaction_satisfied'], y_test['prediction'])
matrix
# 9423 TRUE POSITIVE
# 2202 FALSE POSITIVE
# 2025 TRUE NEGATIVE
# 7069 FALSE NEGATIVE

```
